```python
import pandas as pd
import matplotlib.pyplot as plt

# 바로 읽혀요!
df = pd.read_csv('data/uci-secom.csv')

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist()[:10], "...")
print("\n결측값 확인:")
print(df.isnull().sum().sum(), "개 결측값")
df.head()
```

```python
import matplotlib.pyplot as plt

# Pass/Fail 분포 확인
print("=== Pass/Fail 분포 ===")
print(df['Pass/Fail'].value_counts())
print(f"\n불량률: {(df['Pass/Fail'] == 1).sum() / len(df) * 100:.1f}%")
print(f"정상률: {(df['Pass/Fail'] == -1).sum() / len(df) * 100:.1f}%")

# 시각화
plt.figure(figsize=(6, 4))
df['Pass/Fail'].value_counts().plot(kind='bar', color=['red', 'steelblue'], edgecolor='black')
plt.title('SECOM Pass/Fail 분포', fontsize=14)
plt.xlabel('결과 (-1=정상, 1=불량)')
plt.ylabel('개수')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
